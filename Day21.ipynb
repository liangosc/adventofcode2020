```python
f = open('day21_input.txt')
lines = f.readlines()
f.close()
```

```python
data_raw = [(line[0:-1]) for line in lines]
print(data_raw)
```

```python
len(data_raw)
```

```python
a_list = []
i_list = []
for data in data_raw:
    p1, p2 = data.split('(')
    ingrs = [ing.strip() for ing in p1.split(' ')]
    if ingrs[-1] == '':
        ingrs = ingrs[0:-1]
    
    algs = [alg.strip() for alg in p2[0:-1].split(',')]
    algs[0] = algs[0].split(' ')[1]
    a_list.append(algs)
    i_list.append(ingrs)
    print(algs)
    print(ingrs)
    print('...............')
```

```python
a_list
```

```python
a_list_1d = []
for tmp in a_list:
    for thing in tmp:
        a_list_1d.append(thing)
```

```python
set(a_list_1d)
```

```python
d0 = {}
for item in set(a_list_1d):
    result1 = []
    for i in range(len(a_list)):
        if item in a_list[i]:
            result1.append(i)
    print(f'for {item}:')
    print(result1)
    
    ans = i_list[result1[0]]
    for i in range(1, len(result1)):
        ans = set(ans).intersection(i_list[result1[i]])
    print(ans)
    d0[item] = ans
    print('...........')
```
