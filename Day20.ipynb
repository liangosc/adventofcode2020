```python
f = open('day20_input.txt')
lines = f.readlines()
f.close()
```

```python
data_raw = [(line[0:-1]) for line in lines]
print(data_raw)
```

```python
all_tiles = {}
ct = 0
for data in data_raw:
    if data.startswith('Tile'):
        _, key = data[0:-1].split(' ')
        if key in all_tiles:
            print(f'{key} already there.')
        else:
#             print(key)
            current_key = int(key)
            all_tiles[current_key] = []
    elif data == '':
        continue
    else:
        all_tiles[current_key].append(data)
```

```python
def get_all_edges_from_tile(tile: list) -> list:
    top = tile[0]
    left = ''.join([row[0] for row in tile])
    right = ''.join([row[-1] for row in tile])
    bottom = tile[-1]
    return [top, top[::-1], left, left[::-1], bottom, bottom[::-1], right, right[::-1]]
```

```python
all_edges = dict.fromkeys(all_tiles)
for key, tile in all_tiles.items():
    all_edges[key] = get_all_edges_from_tile(tile)
```

```python
def find_matching_tiles(tile_num: int, side: int, tiles: dict, edges: dict) -> None:
    current_tile = tiles[tile_num]
    target_edge = all_edges[tile_num][side]
    for key, edges in all_edges.items():
        if target_edge in edges:
            if key != tile_num:
                print(f'{tile_num} is next to {key}')
                next_tile = key
                matching_side = edges.index(target_edge)
                new_target_side = (matching_side + 4) % 8
                find_matching_tiles(next_tile, new_target_side, tiles, edges)
```
